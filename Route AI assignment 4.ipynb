```python
import numpy as np

# Create a 2D array with shape (3, 3) and random integers between 1 and 20
random_array = np.random.randint(1, 21, size=(3, 3))

random_array
```

```python
# Create a 5x5 array with random values
random_array = np.random.random((5, 5))

print("Original Array:")
print(random_array)

# Find the minimum and maximum values for each row
min_values = np.min(random_array, axis=1)
max_values = np.max(random_array, axis=1)

print("\nMinimum values for each row:")
print(min_values)

print("\nMaximum values for each row:")
print(max_values)
```

```python
# Create a random vector of size 10
random_vector = np.random.random(10)

print("Original Vector:")
print(random_vector)

# Sort the vector
sorted_vector = np.sort(random_vector)

print("\nSorted Vector:")
print(sorted_vector)
```

```python
# Create an example array
arr = np.array([3, 1, 4, 2, 2, 1, 3, 4, 4, 5, 1, 2, 4])

# Find the most frequent value
most_frequent_value = np.bincount(arr).argmax()

print("Array:")
print(arr)

print("\nMost Frequent Value:")
print(most_frequent_value)
```

```python
# Create a 3x3 array with random values
random_array = np.random.random((3, 3))

print("Original Array:")
print(random_array)

# Subtract the mean of each row from each element
mean_subtracted_array = random_array - np.mean(random_array, axis=1, keepdims=True)

print("\nArray after Subtracting Mean of Each Row:")
print(mean_subtracted_array)
```

```python
# Create a 1D array with the first 20 odd numbers
odd_numbers = np.arange(1, 41, step=2)[:20]

print("Array of the First 20 Odd Numbers:")
print(odd_numbers)
```

```python
# Create a 5x5 array with random values
random_array = np.random.random((5, 5))

print("Original Array:")
print(random_array)

# Find the second-largest value in each column
second_largest_values = np.partition(random_array, -2, axis=0)[-2, :]

print("\nSecond-Largest Value in Each Column:")
print(second_largest_values)
```

```python
# Create a 4x4 array with random values
random_array = np.random.random((4, 4))

print("Original Array:")
print(random_array)

# Find the sum of each row
row_sums = np.sum(random_array, axis=1)

print("\nSum of Each Row:")
print(row_sums)
```

```python
import pandas as pd

# Task 1: Load CSV file into a Pandas DataFrame
file_path = "D:/Route AI & ML/Datasets/data.csv"
df = pd.read_csv(file_path)
```

```python
# Display the first 5 rows of the DataFrame
df.head()
```

```python
df.info()
```

```python
# Display summary statistics for numerical columns
print(df.describe())
```

```python
# Assuming 'Column1' and 'Column2' are the existing columns
df['Total'] = df['TotalPay'] + df['TotalPayBenefits']

# Display the updated DataFrame with the new 'Total' column
df.head()
```

```python
#Condition
condition = df['Total'] > 120

filtered_df = df[condition]

filtered_df.head()
```

```python
#Sorting
sorted_df = df.sort_values(by='Total')
sorted_df.head()
```

```python
#Calculate Average
average_value = df['Total'].mean()

print(f"The average value of 'Total' is: {average_value}")
```

```python
# Most common
most_frequent_value = df['JobTitle'].mode().iloc[0]

print(f"The most frequent value in 'JobTitle' is: {most_frequent_value}")
```

```python
# Assuming 'CategoryColumn' is the categorical column
# and 'NumericColumn' is the numerical column for which you want to calculate the mean
grouped_df = df.groupby('JobTitle').agg({'Total': 'mean'})

# Display the DataFrame with mean values for each group
grouped_df.head()
```

```python
import matplotlib.pyplot as plt

plt.hist(df['Total'], bins=10, color='blue', edgecolor='black')

plt.xlabel('Total')
plt.ylabel('Frequency')
plt.title('Histogram of NumericColumn')

plt.show()
```

```python
category_counts = df['Status'].value_counts()

category_counts.plot(kind='bar', color='skyblue', edgecolor='black')

plt.xlabel('Status')
plt.ylabel('Frequency')
plt.title('Bar Chart of CategoryColumn')

plt.show()
```

```python
import pandas as pd

df['Total'].fillna(df['Total'].mean(), inplace=True)

df['Status'].fillna(df['Status'].mode().iloc[0], inplace=True)

df.head()
```

```python
import pandas as pd
import numpy as np
np.random.seed(42)

#  df1
df1 = pd.DataFrame({
    'common_column': np.random.randint(1, 10, size=4),
    'data1': np.random.choice(['A', 'B', 'C', 'D'], size=4)
})


print("DataFrame df1:")
df1.head()
```

```python
# df2
df2 = pd.DataFrame({
    'common_column': np.random.randint(1, 10, size=4),
    'data2': np.random.choice(['X', 'Y', 'Z', 'W'], size=4)
})

print("\nDataFrame df2:")
df2.head()
```

```python
# Merged
merged_df = pd.merge(df1, df2, on='common_column', how='inner')
print("\nMerged DataFrame:")
merged_df.head()
```
